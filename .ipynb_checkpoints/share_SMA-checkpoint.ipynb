```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def openTradeHistory():
    with open('data02/trade_history02.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["time", "type", "current", "reward", "risk", "close"])
        file.close()
```

```python
def createTradeHistory(trade_detail):
    with open('data02/trade_history02.csv', 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(trade_detail)
        file.close()
```

```python
def closeTrade(current_price):
    df = pd.read_csv('data02/trade_history02.csv')
    for index, row in df.iterrows():
        if row['close'] != 0:
            continue
        if(row['reward'] <= current_price or row['risk'] >= current_price):
            df.loc[index, 'close'] = current_price
            
    df.to_csv('data02/trade_history02.csv', index=False)
```

```python
stockData = pd.Series([],dtype='float64') 
original_data = pd.read_csv('data02/data_set02.csv')
ticker = original_data.iloc[:, 1]

plt.plot(ticker)
```

```python
plt.figure(figsize=(20, 10))
plt.plot(ticker, c='y')

openTradeHistory()
for val in ticker:
    closeTrade(val)
    stockData[len(stockData)] = val
    
    if len(stockData) > 10:
        shortWindow = stockData.rolling(window=6, min_periods=1).mean()
        longWindow = stockData.rolling(window=10, min_periods=1).mean()
        idx = len(stockData) - 1
        
        if shortWindow[idx-1] < longWindow[idx-1] and shortWindow[idx] >= longWindow[idx]:
            plt.plot(idx, val, c='r', marker='o') #Long
            createTradeHistory([idx, "Long", val, (val+1.0), (val-1.0), 0])
    
        #if shortWindow[idx-1] > longWindow[idx-1] and shortWindow[idx] < longWindow[idx]:
            #plt.plot(idx, val, c='g', marker='o') #short
            
plt.plot(shortWindow, c='b')
plt.plot(longWindow, c='r')
```
