```python
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker

import talib
```

```python
input_data = 'data01/SBI_4_5_2021_4_5_2020.csv'
```

```python
stockData = pd.Series([],dtype='float64') 
original_data = pd.read_csv(input_data)
ticker = original_data.iloc[:600, 3]
```

```python
shortWindow = talib.EMA(ticker, timeperiod=8)


plt.figure(figsize=(20, 10))
plt.plot(ticker, c='y')
plt.plot(shortWindow, c='g')

plt.plot(10, ticker[10], c='r', marker='o')

plt.plot(37, ticker[37], c='r', marker='o')

plt.plot(48, ticker[48], c='r', marker='o')
```

```python
plt.figure(figsize=(20, 10))
plt.plot(ticker, c='y')
plt.plot(shortWindow, c='g')


i = 50
begining = False
start_index = 0
while i < len(shortWindow):
    if shortWindow[i] > shortWindow[i-1] and not(begining):
        #print('index {} = {}, index {} = {}'.format(i, shortWindow[i], i-1, shortWindow[i-1]))
        begining = True
        start_index = i
    elif shortWindow[i] < shortWindow[i-1] and begining:
        #print('index {} = {}, index {} = {}'.format(i, shortWindow[i], i-1, shortWindow[i-1]))
        #print('-------------------')
        if (shortWindow[i] - shortWindow[start_index+4]) > 0.75:
            plt.plot(start_index, ticker[start_index], c='r', marker='o')
            plt.plot(start_index+1, ticker[start_index+1], c='r', marker='o')
            plt.plot(start_index+2, ticker[start_index+2], c='r', marker='o')
        begining = False
    i = i+1
```
